### 4.Read the given data set and answer the following questions. Show the steps taken to produce your final answer.

1. Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.
2. How many patients in the data are having the blood pressure level above 120?
3. Find the minimum and maximum value of insulin level in the data.
4. Create a histogram of "Age".
5. Create a scatter plot of "BMI" vs "Glucose".

```python
diabetes = pd.read_csv('diabetes_model.csv')
pd.set_option('display.max_rows',614)
diabetes
```

### 1.Glucose level & Age for the samples having the index ranging from 45 to 60.

```python
data1 = diabetes[["Glucose","Age"]].iloc[45:61]
data1
```

### 2.patients in the data having the blood pressure level above 120.

```python
data2 = diabetes[diabetes["BloodPressure"]>120]
data2
```

Only 1 patient have BP above 120.

### 3.Min and Max of insulin level.

```python
data3 = diabetes["Insulin"].max()
data4 = diabetes["Insulin"].min()
print("Minimum value of Insulin level in the data is",data4)
print("Maximum value of Insulin level in the data is",data3)
```

### 4. Histogram of Age.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,5))
plt.title("Histogram of Age")
plt.hist(diabetes["Age"],rwidth=0.9)
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid()
```

### 5. Scatter plot of "BMI" vs "Glucose".

```python
plt.figure(figsize=(10,7))

plt.scatter(x="BMI",y="Glucose",data=diabetes,s=15)
plt.title("BMI vs Glucose")
plt.xlabel('BMI')
plt.ylabel('Glucose')
```
